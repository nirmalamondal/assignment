```python
#Q2. What is the objective function of a linear SVM?
"""
The objective function of a linear Support Vector Machine (SVM) is to find the optimal hyperplane that maximizes the margin between the two
classes while correctly classifying the training data.
The objective of the SVM is to find the optimal values of the weight vector w and the bias term b that satisfy the margin constraint while 
minimizing the regularization term. This is typically done using optimization algorithms that can handle convex quadratic programming 
problems, such as the Sequential Minimal Optimization (SMO) algorithm or other gradient-based optimization methods. Once the optimization 
problem is solved, the resulting weight vector w and bias term b define the hyperplane, and the decision function can be used to classify
new data points.
"""
```

```python
#Q3. What is the kernel trick in SVM?
"""
The kernel trick is a technique used in Support Vector Machines (SVM) to handle nonlinearly separable data by implicitly transforming the 
original feature space into a higher-dimensional feature space. It allows SVM to find a linear decision boundary in this higher-dimensional
space, even though the original data might not be linearly separable in its original feature space. This is achieved without explicitly 
computing the transformation, thus avoiding the computational cost and memory requirements associated with working in high-dimensional 
spaces.

In SVM, the goal is to find the hyperplane that maximizes the margin between the two classes while correctly classifying the data points. 
In the case of linearly separable data, a simple linear hyperplane can be used to separate the classes. However, real-world data is often
nonlinearly separable, and finding a linear hyperplane in the original feature space may not be sufficient.

The kernel trick addresses this issue by introducing a kernel function (also known as a Mercer kernel) that calculates the dot product of 
the data points in the higher-dimensional feature space without explicitly transforming the data into that space. This kernel function 
essentially measures the similarity between two data points in the higher-dimensional space, allowing SVM to work with the data in this
implicit high-dimensional feature space.
"""
```

```python
#Q4. What is the role of support vectors in SVM Explain with example
"""
The role of support vectors in Support Vector Machines (SVM) is crucial for defining the decision boundary and maximizing the margin between
the classes. Support vectors are the data points that lie closest to the decision boundary (hyperplane) and have the most influence on the
positioning and orientation of the decision boundary.

In a binary SVM, the goal is to find the hyperplane that separates two classes of data points with the largest margin. This hyperplane is 
determined by a subset of training data points, known as support vectors. The hyperplane is positioned in a way that it stays equidistant 
from the closest support vectors of both classes. The margin is defined as the perpendicular distance from the hyperplane to the support 
vectors.

Here's an example to illustrate the role of support vectors in SVM:
Suppose we have a simple 2D dataset with two classes, represented by different colors:

Class A (positive): X1 = [(1, 2), (2, 3), (3, 3)]
Class B (negative): X2 = [(3, 6), (4, 7), (5, 5)]

In this 2D space, the SVM aims to find a decision boundary (hyperplane) that separates the two classes with the largest margin.

In this example, let's assume that the SVM finds the following decision boundary:

Equation of the decision boundary: w Â· x + b = 0

where:

w is the weight vector perpendicular to the hyperplane.
x is the input feature vector (x, y) in 2D space.
b is the bias term that shifts the hyperplane.
Now, let's visualize the decision boundary along with the support vectors:

In this example, the support vectors are the data points that lie closest to the decision boundary. In this case, the support vectors are:

(1, 2) from Class A
(5, 5) from Class B
(4, 7) from Class B
The support vectors play a critical role in defining the decision boundary because they dictate the position and orientation of the
hyperplane. If we were to remove any non-support vector data points, the position of the hyperplane would remain unchanged. However, if we 
were to remove any support vector, the hyperplane would shift, and the margin would be affected.

In summary, support vectors are the key data points that determine the decision boundary and the margin in SVM. They are essential for 
generalization and play a significant role in constructing an effective and robust classifier. SVM focuses on these critical points to make 
a clear separation between the classes and avoid overfitting to noisy or less informative data points.
"""
```

```python
#Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?
"""Hyperplane:
The hyperplane is the decision boundary that separates the data points of different classes in the feature space. In a 2D dataset, the 
hyperplane is a straight line. The goal of the SVM is to find the hyperplane that maximizes the margin between the two classes. 
Marginal Plane:
The marginal planes are parallel to the hyperplane and are positioned at the margins, defined by the closest support vectors. The margins
are the perpendicular distances from the hyperplane to the support vectors. 
Soft Margin:
In some cases, the data may not be perfectly separable, or there might be outliers that prevent finding a strict hyperplane. Soft Margin
SVM allows some data points to violate the margin or even be misclassified to achieve a more flexible decision boundary. 

Hard Margin vs. Soft Margin:
In the hard margin SVM, the decision boundary is rigid and strictly separates the classes, which may lead to overfitting if there are noisy 
data points or slight variations in the data. Soft margin SVM, on the other hand, allows some flexibility in the decision boundary, leading 
to better generalization and robustness to outliers or noisy data. The choice of the regularization parameter C determines the trade-off 
between maximizing the margin and minimizing classification errors or violations of the margin.
"""
```

```python
"""
SVM Implementation through Iris dataset.
Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
~ Compute the accuracy of the model on the testing setl
~ Plot the decision boundaries of the trained model using two of the featuresl
~ Try different values of the regularisation parameter C and see how it affects the performance of
the model.
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Step 1: Load the Iris dataset
iris = load_iris()
X, y = iris.data[:, :2], iris.target

# Step 2: Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 3: Train a linear SVM classifier on the training set
clf = SVC(kernel='linear')

# Step 4: Train the model and predict the labels for the testing set
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# Step 5: Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Step 6: Plot the decision boundaries of the trained model using two of the features
def plot_decision_boundary(X, y, clf, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Sepal Length (cm)')
    plt.ylabel('Sepal Width (cm)')
    plt.title(title)
    plt.show()

plot_decision_boundary(X_train, y_train, clf, 'Decision Boundary (Training Set)')
plot_decision_boundary(X_test, y_test, clf, 'Decision Boundary (Testing Set)')





C_values = [0.1, 1, 10, 100, 1000]

for C_val in C_values:
    clf = SVC(kernel='linear', C=C_val)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"C = {C_val}, Accuracy: {accuracy:.2f}")
```

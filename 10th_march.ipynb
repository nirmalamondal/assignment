```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

'''Estimation statistics is a branch of statistics that deals with estimating unknown parameters of a population based on sample data.
Point Estimate:
A point estimate is a single value that is used to estimate the unknown parameter of the population. It's obtained from the sample data and serves as 
the best guess for the true value of the parameter.
Interval Estimate:
An interval estimate provides a range of values within which the true population parameter is likely to lie, along with a level of confidence 
associated with that range. It's created by using a point estimate and calculating an interval or range around this estimate.
'''



Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

'''Hypothesis testing is a fundamental statistical method used to make decisions or draw conclusions about a population based on sample data.

Importance of Hypothesis Testing:
1.Inference and Decision Making
2.Scientific Validity
3.Business and Industry Applications
4.Policy and Decision-Making
5.Quality Control
6.Continuous Improvement

'''
Q4.Create a hypothesis that states whether the average weight of male college students is greater than the average weight of female college students.

'''Null Hypothesis (H0): The average weight of male college students is equal to or less than the average weight of female college students.
Alternative Hypothesis (H1): The average weight of male college students is greater than the average weight of female college students.'''

Q6: What is a null and alternative hypothesis? Give some examples.

'''Null Hypothesis:The null hypothesis represents the default assumption or a statement of no effect, no difference, or no relationship in the 
population.It is typically the hypothesis being tested, aiming to be rejected if there is sufficient evidence against it.Symbolically, the null
hypothesis is denoted as H0
Alternate Hypothsis:The alternative hypothesis contradicts the null hypothesis and represents the claim or effect that is being sought or tested for.
It is the hypothesis that researchers want to support if there is enough evidence to reject the null hypothesis.Symbolically, the alternative 
hypothesis is denoted as H1.
Examples of null and alternative hypotheses :
Scenario: Sales of two company named as A and B.
Null Hypothesis : The average sales with Advertisement A are less than or equal to the average sales with Advertisement B.
Alternative Hypothesis : The average sales with Advertisement A are greater than the average sales with Advertisement B.
'''



Q7: Write down the steps involved in hypothesis testing.

'''1.State the null hypothesis and the alternative hypothesis 
2.Choose a significance level that determines the threshold for rejecting the null hypothesis. Common values include 0.05 or 0.01
3.Choose an appropriate test statistic (e.g., z-test, t-test, chi-square test) based on the type of data, sample size, and the hypothesis being tested.
4.Gather and collect data from a random sample that is representative of the population of interest.
5.Calculate the test statistic using the collected sample data. The formula for the test statistic depends on the chosen statistical test.
6.Based on the test statistic and chosen significance level, determine the critical region or calculate the p-value associated with the test.
7.Compare the calculated test statistic to the critical value from the distribution or compare the p-value to the significance level 
8.If the test statistic falls into the critical region or if the p-value is less than α, reject the null hypothesis. Otherwise, fail to reject the
null hypothesis.
9.If the null hypothesis is rejected, support the alternative hypothesis. If the null hypothesis is not rejected, there is insufficient evidence to 
support the alternative hypothesis.'''


Q8. Define p-value and explain its significance in hypothesis testing.

'''
The p-value, or probability value, is a crucial concept in hypothesis testing that quantifies the strength of evidence against the null hypothesis(H0).
Significance:
The p-value serves as a measure to determine the strength of evidence against the null hypothesis in hypothesis testing. It helps researchers make
decisions about rejecting or failing to reject the null hypothesis based on the observed sample data and the chosen significance level.
'''

Q11: What is Student’s t distribution? When to use the t-Distribution.
'''

The Student's t-distribution (often referred to simply as the t-distribution) is a probability distribution that arises in statistics and is used in 
hypothesis testing, confidence interval estimation, and when dealing with small sample sizes. It's similar in shape to the standard normal
distribution (bell-shaped curve), but with heavier tails.

When to use the t-distribution:
1.Small Sample Sizes
2.Estimation and Inference
3.Hypothesis Testing 
4.Confidence Intervals'''
Q12: What is t-statistic? State the formula for t-statistic.

'''
The t-statistic is a statistical measure that quantifies the difference between sample means (or between a sample mean and a population mean) in 
standard deviation units.
t= (Sample mean - population mean)/( sample standard deviation/root n)
'''
Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean 
revenue to be $500 with a standard deviation of $50. Estimate the population mean revenue with a 95% confidence interval.

'''
confidence Interval = x + Z * (s/root  n)
                        -
                    = 500 + 1.96 * (50/ root 50)
                          -
Lower Limit =500 − 1.96 * 7.07 ≈ 485.14
Upper Limit =500 + 1.96 * 7.07 ≈ 514.86  
                        
'''
Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that
the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.

'''
To test the researcher's hypothesis regarding the decrease in blood pressure using the provided sample data, we can conduct a one-sample z-test.
The null hypothesis states that the new drug does not decrease blood pressure by 10 mmHg.
The alternative hypothesis contradicts the null hypothesis, suggesting that the new drug decreases blood pressure by an amount different from 10 mmHg.
x is the sample mean (8 mmHg)
μ is the hypothesized population mean (10 mmHg)
s is the sample standard deviation (3 mmHg)
n is the sample size (100)
Let's calculate the z-statistic:
z=( 8-10)/(3/10) ≈ −6.67
For a sample size of 100 and a significance level of 0.05, the critical z-value is approximately ±1.984.
Since the calculated z-statistic (-6.67) is well beyond the critical values (-1.984 and +1.984), we reject the null hypothesis .
'''
Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a standard deviation of 0.5 pounds. A random sample
of 25 products is taken, and the sample mean weight is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less 
than 5 pounds with a significance level of 0.01.


'''
population mean weight = 5 pounds
standard deviation = 0.5 pounds
samplesize = 25 
sample mean = 4.8
degree of freedom = 25 -1 =24
as n=25 which is less than 30 so we consider the t-test.
t = (4.8 - 5)/(0.5/5) = -2
For a one-tailed test and a significance level of 0.01, the critical t-value is approximately -2.492.
Since the calculated t-statistic (-2) is smaller (more extreme) than the critical t-value (-2.492), it falls in the rejection region.

'''
Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =30) has a mean score of 80 with a standard
deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means
for the two groups are equal with a significance level of 0.01.

'''To test the hypothesis that the population means for the two groups are equal, you can use a two-sample t-test for independent samples. This test 
compares the means of two independent groups to determine if there is a statistically significant difference between them.

Given:

Group 1: n1=30, x1=80,s1=10
Group 2: n2=40,x2=75,s2=8
The null hypothesis for the two-sample t-test is that the population means of the two groups are equal
The alternative hypothesis is that the population means of the two groups are not equal

The two-sample t-test formula is: (x1-x2)/(root of((s1^2/n1) + (s2^2/n2)))  
                                 = (80-75)/root of(10^2/30 + 8^2/40) = 2.25
                                 
 you can find the critical t-value using statistical tables or software.
Compare the calculated t-value to the critical t-value. If the calculated t-value exceeds the critical t-value (in either the positive or negative
direction), you reject the null hypothesis and conclude that there is a statistically significant difference between the means of the two groups.
Performing the calculations will give you the t-value and allow you to determine whether to reject the null hypothesis based on the significance
level chosen (0.01) and the degrees of freedom.
'''
Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers
and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.


'''
To estimate the population mean number of ads watched by viewers during a TV program with a 99% confidence interval based on the provided sample data,
you can use the formula for a confidence interval for the population mean when the population standard deviation is unknown:

Confidence Interval= x ± t×( n *s/root of n)
Where:
x is the sample mean number of ads watched (4).
s is the sample standard deviation (1.5).

n is the sample size (50).
t is the critical value from the t-distribution corresponding to the desired confidence level and degrees of freedom (n−1) = 50 - 1 = 49
First, find the critical t-value for a 99% confidence level and degrees of freedom 
You can use statistical tables or software to determine this value.

Next, calculate Margin of Error= Critical t-value×(1.5/root of 50)

Finally, calculate the confidence interval:
Confidence Interval
=x ± Margin of Error

Perform the calculations to find the margin of error and then compute the confidence interval using the sample mean and the margin of error.

This will provide you with a range within which you can be 99% confident that the true population mean number of ads watched by viewers during the TV
program lies based on the sample data.'''
```

```python
'Q5. Write a Python script to conduct a hypothesis test on the difference between two population means, given a sample from each population.'
from scipy import stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)
    if p_value < alpha:
        print(f"Reject the null hypothesis. There is a significant difference between the population means.")
    else:
        print("Fail to reject the null hypothesis. There is no significant difference between the population means.")

    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

sample1 = [78, 82, 75, 88, 92]
sample2 = [72, 79, 85, 81, 77]

two_sample_t_test(sample1, sample2)
```

```python
"Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10."

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t

degrees_of_freedom = 10
x = np.linspace(-4, 4, 1000)  # Range of x values
pdf_values = t.pdf(x, df=degrees_of_freedom)

# Plot the t-distribution
plt.figure(figsize=(4,4))
plt.plot(x, pdf_values, label=f"t-distribution (df={degrees_of_freedom})")
plt.title(f"Student's t-Distribution (Degrees of Freedom = {degrees_of_freedom})")
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

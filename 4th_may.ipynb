```python
Q1. What is a time series, and what are some common applications of time series analysis?
'''
Time series: A sequence of data points indexed in time order, typically equally spaced.
Applications: Forecasting stock prices, weather forecasting, predicting sales, analyzing economic trends, monitoring sensor data, studying medical
data trends, etc.
'''
Q2. What are some common time series patterns, and how can they be identified and interpreted?
'''
Common patterns: Trend, Seasonality, Cyclical, and Irregular (Noise).

Identification and interpretation:
1.Trend: Long-term movement direction of data.
2.Seasonality: Regular fluctuations or patterns.
3.Cyclical: Repeating up-and-down movements without fixed periods.
4.Irregular: Random variations in the data.
'''
Q3. How can time series data be preprocessed before applying analysis techniques?
'''
Preprocessing techniques: 
Handling missing values, smoothing data, detrending, deseasonalizing, transforming (logarithm, differencing), 
and handling outliers.
'''
Q4. How can time series forecasting be used in business decision-making, and what are some common challenges and limitations?
'''
Business decision-making: Forecasting helps in inventory management, workforce planning, budgeting, and strategic decision-making.
Challenges: Short-term fluctuations, unexpected events, noisy data, overfitting, and model accuracy in dynamic environments.
''''
Q5. What is ARIMA modelling, and how can it be used to forecast time series data?
'''
ARIMA (AutoRegressive Integrated Moving Average): A model used for time series forecasting that combines autoregressive, differencing, and moving 
average components to capture different patterns.
Usage: ARIMA is applied to time series data after identifying and removing trends or seasonality.
'''
Q6. How do Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots help in identifying the order of ARIMA models?
'''
ACF: Shows the correlation between a series and its lagged values.
PACF: Reflects the correlation between a series and a lagged version of itself after removing the effects of intervening time lags.
Identification: Patterns in ACF and PACF help determine the orders of ARIMA models (p, d, q).
'''
Q7. What are the assumptions of ARIMA models, and how can they be tested for in practice?
'''
Assumptions: Stationarity, no autocorrelation in residuals, normality of residuals.
Testing: Statistical tests for stationarity (e.g., ADF test), residual analysis (e.g., Ljung-Box test), and checking residual normality(QQ plots).
'''
Q8. Suppose you have monthly sales data for a retail store for the past three years. Which type of time series model would you recommend for 
forecasting future sales, and why?
'''
ARIMA or Seasonal ARIMA may be suitable for forecasting sales data as it can capture both trend and seasonality, common in retail sales.
'''
Q9. What are some of the limitations of time series analysis? Provide an example of a scenario where the limitations of time series analysis may be
particularly relevant.
'''
Limitations: Inability to capture sudden, unexpected events (e.g., natural disasters, sudden market changes), extrapolation beyond observed data range.
Scenario: Forecasting sales during a pandemic where consumer behavior drastically changes due to unforeseen circumstances.
'''
Q10. Explain the difference between a stationary and non-stationary time series. How does the stationarity of a time series affect the choice of 
forecasting model?
'''
Stationary series: Mean, variance, and covariance are constant over time.
Non-stationary series: Mean, variance, or covariance change over time.
Effect on forecasting model: Stationarity is crucial for ARIMA models. For non-stationary series, transformations or differencing may be applied to
achieve stationarity before applying ARIMA. Other models like exponential smoothing might be used for non-stationary data.
'''
```

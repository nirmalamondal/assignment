```python
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain withan example.
'''The PMF is a function that describes the probability distribution of a discrete random variable. It gives the probability that the random variable takes on a specific value. For discrete random variables, the PMF is represented by a table, graph, or formula that assigns probabilities to each possible value the variable can take.
Consider a fair six-sided die. The random variable X represents the outcome of rolling the die. The PMF of this random variable is given by:
P(X=x)= 1/6, where x=1,2,3,4,5,6


The PDF is a function used to describe the probability distribution of a continuous random variable. It represents the likelihood of the variable 
taking on a particular value within a range. Unlike the PMF for discrete variables, the PDF doesn't give the probability at a specific value but 
rather the probability density at that value.
Consider a continuous random variable Y representing the heights of adult individuals. Let's assume the heights follow a normal distribution with a 
mean of 170 cm and a standard deviation of 10 cm. 
The PDF of the normal distribution for heights shows the relative likelihood of observing different heights. For example, it indicates that heights
closer to the mean (170 cm) are more likely to occur than those farther away.
'''
Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
'''The Cumulative Distribution Function (CDF) is a function that describes the probability that a random variable takes on a value less than or equal 
to a given value. It's applicable to both discrete and continuous random variables.
Consider a continuous random variable Y representing the heights of adults following a normal distribution with a mean of 170 cm and a standard
deviation of 10 cm. The CDF F(y) at a specific height, say y=180 cm, would indicate the probability that a randomly selected individual has a height 
less than or equal to 180 cm.

uses:
Describing cumulative probabilities: CDF provides a way to describe the cumulative probabilities of a random variable taking on values up to a 
certain point.
Understanding distributions: It helps in understanding the spread, range, and probabilities associated with different values of a random variable.
Statistical analysis: CDFs are used in statistical analysis, hypothesis testing, confidence interval calculations, and modeling various phenomena in 
different fields like finance, engineering, and sciences.'''
Q3: What are some examples of situations where the normal distribution might be used as a model?
'''
1.Biological Measurements
2.Financial Markets
3.Natural Phenomena
4.Psychometrics
5.Quality Control
'''
Explain how the parameters of the normal distribution relate to the shape of the distribution.
''' the mean determines the central position and symmetry of the normal distribution, while the standard deviation controls the spread or variability
of the distribution. Adjusting these parameters results in shifting, stretching, or compressing the distribution while maintaining its characteristic 
bell-shaped curve.
'''
Q4: Explain the importance of Normal Distribution. Give a few real-life examples of NormalDistribution.
'''Importance:
1.Central Limit Theorem
2.Statistical Analysis and Inference
3.Modeling Real-world Phenomena


Real-life examples:
1.Biological Measurements
2.Financial Markets
3.Natural Phenomena
4.Psychometrics
5.Quality Control
'''
Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?
'''The Bernoulli distribution is a discrete probability distribution that models a random experiment with two possible outcomes: success (typically
denoted as 1) and failure (typically denoted as 0). It is named after Swiss mathematician Jacob Bernoulli.

Consider a single coin flip. Let X be a random variable that represents the outcome of the coin flip, where getting heads is considered a success (1) 
and getting tails is considered a failure (0). If the probability of getting heads (p) is 0.5 for a fair coin, then X follows a Bernoulli distribution
with p=0.5. The PMF would be P(X=1)=0.5 for heads and P(X=0)=0.5 for tails.

DIFFERENCES:
1.Number of Trials:
Bernoulli Distribution: Represents a single trial (one event).
Binomial Distribution: Represents the number of successes in a fixed number of independent Bernoulli trials.
2.Parameters:
Bernoulli Distribution: Has a single parameter 
Binomial Distribution: Has two parameters 
3.Random Variable:
Bernoulli Distribution: Describes the outcome of a single trial.
Binomial Distribution: Describes the total number of successes in 
'''
Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset is normally distributed, what is the probability
that a randomly selected observation will be greater than 60? Use the appropriate formula and show your calculations.
''' mean = 50, std = 10,X=60
as X>=30  and mean is given then we can apply ztest
z-score = (X-mean)/std
        = (60 - 50)/10
        =1
For a Z-score of 1( by referring to a Z-table), the probability of being greater than 60 is approximately 0.1587 or 15.87%.
'''

Q7: Explain uniform Distribution with an example.
'''The Uniform Distribution is a probability distribution where every value within a certain range is equally likely to occur. In other words, each 
value has the same probability of occurring, and the probability density function (PDF) is constant within the range.
Example of Uniform Distribution:

Consider a six-sided fair die. The outcome of rolling the die follows a discrete uniform distribution between 1 and 6, where each side has an equal
probability of 1/6 of occuring
For a continuous example, imagine a scenario where a spinner is equally likely to stop anywhere between 0 and 1. The angle at which the spinner stops
represents a continuous uniform distribution between 0 and 1. Any value between 0 and 1 has an equal chance of being the result.'''
Q8: What is the z score? State the importance of the z score.
'''
The z-score, also known as the standard score, is a statistical measure that quantifies the number of standard deviations a data point is from the
mean of a dataset. It standardizes individual data points within a distribution, allowing comparisons between different distributions with different
means and standard deviations.

IMPORTANCE:
1.Standardization and Comparison
2.Identifying Outliers
3.Data Normalization
4.Hypothesis Testing
5.Probability Calculations
'''

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
'''The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes the behavior of the distribution of sample means when
drawing repeated samples from a population, regardless of the shape of the original population distribution.


Significance of the Central Limit Theorem:
1.Foundation of Inferential Statistics: The CLT forms the foundation of inferential statistics. It allows statisticians to make inferences about 
population parameters based on sample statistics, even when the population distribution is unknown or non-normally distributed.

2.Statistical Inference: It enables the use of parametric statistical techniques (e.g., hypothesis testing, confidence intervals) that rely on the
assumption of normality, even for data that may not follow a normal distribution, as long as the sample size is sufficiently large.

3.Real-world Applications: The CLT is widely applicable across various fields such as finance, economics, biology, psychology, and quality control.
It aids in understanding the behavior of sample means and sums in large datasets.
'''
Q10: State the assumptions of the Central Limit Theorem.
'''
The assumptions of the Central Limit Theorem include:

1.Independence 
2.Sample Size
3.Population Distribution
4.Identical Distribution
5.Finite Variance
'''
```

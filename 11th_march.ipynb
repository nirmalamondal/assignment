```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
'''
DIFERENCES:
Sample Size and Assumptions:
t-test: This test is suitable for smaller sample sizes (typically less than 30) when the population standard deviation is unknown and must be 
estimated from the sample data. The t-test assumes that the data are normally distributed.
z-test: Generally, the z-test is applied when the sample size is large (typically greater than 30) and when the population standard deviation is 
known. It assumes that the data follow a normal distribution.

Standardization:
t-test: It uses the sample standard deviation to standardize the difference between sample means.
z-test: It uses the population standard deviation to standardize the difference between sample means.

Distribution:
t-test: The t-distribution is used to determine critical values for the test statistic. The shape of the t-distribution is similar to the normal 
distribution but has heavier tails, especially with smaller sample sizes.
z-test: The z-test uses the standard normal distribution (z-distribution) to determine critical values for the test statistic. The z-distribution has 
a mean of 0 and a standard deviation of 1.


EXAMPLES:
1.t-test - 
Suppose the researcher randomly selects 20 students from each group and measures their heights. The population standard deviation for height is
unknown. In this case, because the sample size is relatively small (less than 30) and the population standard deviation is not given, the researcher 
would use a two-sample t-test for independent samples.

z-test - Now, consider a scenario where a quality control manager wants to assess whether a new manufacturing process improves the mean production 
output of a factory. The manager collects a large sample of 200 products produced using the new process and knows the population standard deviation of
the production output from historical data.
'''
Q2: Differentiate between one-tailed and two-tailed tests.
'''
A one-tailed test, also known as a one-sided test, is used when the hypothesis being tested specifies a direction (either an increase or a decrease) 
in a particular parameter (mean, proportion, etc.).
A two-tailed test, also known as a two-sided test, is used when the hypothesis being tested does not specify a particular direction for the effect.

The critical region (rejection region) for a one-tailed test is only on one side of the distribution curve.
The critical region (rejection region) for a two-tailed test is split between both sides of the distribution curve.
'''
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

'''
A Type I error occurs when the null hypothesis (H0)  is incorrectly rejected when it is actually true. In other words, it's a false positive.
A Type II error occurs when the null hypothesis (H0) is not rejected when it is actually false. It's a false negative.

The null hypothesis (H0) states that the drug has no effect on the disease.
Type I Error Example: Incorrectly concluding that the drug is effective (rejecting H0) when, in fact, the drug has no effect on the disease (the null
hypothesis is true).
Type II Error: Failing to conclude that the drug is effective (failing to reject H0) when, in reality, the drug is effective (the null hypothesis is
false).
'''
 
Q4: Explain Bayes's theorem with an example.
'''
Bayes's theorem is a fundamental concept in probability theory that describes how to update the probability of an event based on new evidence or
information. It calculates the probability of an event occurring given prior knowledge of related conditions. The theorem is expressed mathematically 
as:
P(A∣B)=P(B∣A)×P(A)/ P(B)
Where:
P(A∣B) is the probability of event A occurring given that event B has occurred.
P(B∣A) is the probability of event B occurring given that event A has occurred.
P(A) is the prior probability of event A.
P(B) is the prior probability of event B.
'''
Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
'''
A confidence interval is a statistical range that estimates the true value of a population parameter, such as the mean, proportion, or variance, based
on sample data. It provides a range of values within which the true parameter is likely to fall, along with a degree of confidence associated with
that range.
Confidence Interval for the Mean=Sample Mean ±(Critical Value * Standard Error).

Let's say you want to calculate a 95% confidence interval for the mean height of a population based on a sample of 100 individuals. The sample mean 
height is 65 inches, and the sample standard deviation is 3 inches.

Assuming a normal distribution and using a Z-table for a 95% confidence level (which corresponds to a critical value of approximately 1.96):
Standard Error= 3/root of 100 = 0.3
Confidence Interval=Sample Mean±(Critical Value×Standard Error)=65 ± (1.96 * 0.3) =65 ± 0.588

Therefore, the 95% confidence interval for the mean height of the population is approximately between 
65−0.588=64.412 inches and  65+0.588=65.588 inches.
'''
Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. 
Provide a sample problem and solution.

'''
Suppose a certain genetic mutation occurs in 5% of the population. A test has been developed to detect this mutation, and it has a true positive rate
(sensitivity) of 90% and a true negative rate (specificity) of 85%.
Given this scenario, let's say a person tests positive for the mutation. We want to determine the probability that this person actually has the 
mutation.

Let:
P(M) be the probability of having the mutation (prior knowledge) = 0.05.
P(T∣M) be the probability of testing positive given having the mutation (sensitivity) = 0.90.

P(¬M) be the probability of not having the mutation = 1 - 
P(M) = 0.95.
P(T∣¬M) be the probability of testing positive given not having the mutation (false positive rate) = 1 - specificity = 1 - 0.85 = 0.15.
We want to find 
P(M∣T), the probability of having the mutation given a positive test result, using Bayes' theorem:
P(T)=P(T∣M)×P(M)+P(T∣¬M)×P(¬M)
P(T)=0.90×0.05+0.15×0.95
P(T)=0.045+0.1425
P(T)=0.1875

Now, plug these values into Bayes' theorem:
P(M∣T)= 0.90×0.05/0.1875
P(M∣T)≈0.24

Therefore, the probability that a person actually has the mutation given that they tested positive is approximately 24%.
'''
Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5. Interpret the results.
'''
Confidence Interval=50±(1.96 × (5/root of 100))
Confidence Interval=50±(1.96×5/10) 
Confidence Interval=50±0.98

Therefore, the 95% confidence interval for the population mean based on the given sample is approximately between 
50−0.98=49.02 and 50+0.98=50.98.
'''
Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a
larger sample size would result in a smaller margin of error.
'''

The margin of error in a confidence interval represents the range within which the true population parameter (such as the population mean or
proportion) is likely to lie, considering the variability introduced by sampling. It measures the precision or uncertainty associated with the
estimate obtained from a sample.
Margin of Error=Critical Value × (Standard Deviation/root of(Sample Size))
As the sample size increases the margin of error decreases because the square root of the sample size in the denominator reduces the overall effect on
the margin of error.A larger sample provides more precise estimates, reducing the variability caused by random sampling.

'''
Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the 
results.
'''
Z-score= X−Mean/Standard Deviation
Data point (X) = 75
Population mean = 70
Population standard deviation = 5
Z-score= 75−70/5 =1
Therefore, the z-score for the data point with a value of 75, in relation to a population with a mean of 70 and a standard deviation of 5, is 1.
'''
Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of
2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.
'''
Sample size (n) = 50
Sample mean (x) = 6 pounds
Sample standard deviation (s) = 2.5 pounds
population mean(u) = 0

Null Hypothesis :The weight loss drug is not significantly effective or does not lead to weight loss. 
Alternative Hypothesis : The weight loss drug is significantly effective, resulting in weight loss. 

t-test = x - u/(s/root of n)
       = 6-0/(2.5/root of 50) = 16.95
       
Now, with the calculated t-statistic of approximately 16.95, we compare it with the critical t-value at a 95% confidence level with degrees of freedom
=n−1=49.
Using statistical software or t-tables, the critical t-value for a two-tailed test at a 95% confidence level with 49 degrees of freedom is 
approximately ±2.009.

Since the calculated t-statistic (16.95) is much larger than the critical t-value (2.009), we reject the null hypothesis.       
'''

Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of
people who are satisfied with their job.
'''
confidence interval = 0.65 +- 1.96 * root of(.65 * (1 - .65)/500)
                    = 0.65 +- 1.96 * 0.02134 = 0.65 +- 0.0418
Therefore, the 95% confidence interval for the true proportion of people who are satisfied with their job is approximately between 
0.65−0.0418=0.6082 (or 60.82%) and 0.65+0.0418=0.6918 (or 69.18%).    
'''
Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a 
standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two
teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.
Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval 
for the true population mean.
'''
For a 90% confidence level, the critical value can be obtained from the z-table, which corresponds to a 90% confidence level, typically 
z≈1.645
Confidence Interval = 65 +- (1.645 *(8/root of 50))=65 +- 1.861
Therefore, the 90% confidence interval for the true population mean is approximately between 
65−1.861=63.139 and 65+1.861=66.861.

'''
Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard
deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level
using a t-test.
'''
Null Hypothesis (H0): Caffeine has no significant effect on reaction time.
Alternative Hypothesis (H1): Caffeine has a significant effect on reaction time. 
 
 
To conduct a hypothesis test to determine if caffeine has a significant effect on reaction time, we'll perform a one-sample t-test because we're 
comparing the mean reaction time of a single group to a known or hypothesized value.

Given:
Sample size (n) = 30
Sample mean (x) = 0.25 seconds
Sample standard deviation (s) = 0.05 seconds
Hypothesized mean reaction time (u) - not explicitly given, assuming it's the standard reaction time without caffeine.Since the sample is not compared
to a specific hypothesized value it's not possible to determine the t-statistic directly without knowing the hypothesized mean reaction time.
Assuming a common reaction time without caffeine is 0.30 seconds (just as an example for demonstration purposes)

t = (x-u)/(s/root of n) = (0.25 - 0.30)/(0.05/root of 30) = 5.47

Comparing this calculated t-statistic to the critical t-value (-1.699 and +1.699), we find that the calculated t-statistic is much smaller in 
magnitude than the critical t-value. Therefore, we reject the null hypothesis.
'''
```
